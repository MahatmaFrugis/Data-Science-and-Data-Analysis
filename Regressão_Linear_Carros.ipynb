```python
import seaborn as sns
sns.set(rc = {'figure.figsize':(15,8)})
import numpy as np
import pandas as pd
df_carros = pd.read_csv("carros.csv")
```

```python
df_carros = df_carros.dropna()
```

```python
df_carros.info()
```

```python
min = df_carros["price"].min()
max = df_carros["price"].max()
print(f"preço varia de ${min} a ${max}")
```

```python
upp_mat = np.triu(df_carros.corr())
sns.heatmap(df_carros.corr(), vmin = -1, vmax = +1, linewidths = 0, annot = True, cmap = 'coolwarm', mask = upp_mat)
```

```python
sns.regplot(x = "price", 
            y = "highway mpg", 
            data = df_carros) 
```

```python
from sklearn.linear_model import LinearRegression
X = df_carros[["engine size", "weight", "width"]]
Y = df_carros["price"]
modelo = LinearRegression()
modelo.fit(X,Y)
```

```python
sns.regplot(x = "price", 
            y = "weight", 
            data = df_carros) 
```

```python
y_predito = modelo.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_predito, Y)
RMSE = np.sqrt(MSE)
from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(y_predito, Y)
print(f'MAE:{MAE}\n MSE:{MSE}\n RMSE:{RMSE}')
```
